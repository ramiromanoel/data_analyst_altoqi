```python
import pandas as pd
import re
```

```python
entrada  = r"C:\git\data_analyst_altoqi\python\dados\telefones_poluidos.csv"
saida_csv = r"C:\git\data_analyst_altoqi\python\limpo\telefones_limpos.csv"
saida_excel = r"C:\git\data_analyst_altoqi\python\limpo\telefones_limpos.xlsx"

```

```python
df = pd.read_csv(entrada)
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
def limpar_e_padronizar_telefone(valor):
    if pd.isna(valor):
        return None

    numeros = re.sub(r'\D', '', str(valor))

    if numeros.startswith('0'):
        numeros = numeros[1:]

    n = len(numeros)

    if n < 10:
        return None
    elif n == 10:
        return numeros 
    elif n == 11:
        return numeros
    elif n == 12:
        return numeros[-10:] 
    else: 
        return numeros[-11:]
```

```python
df["telefone_limpo"] = df["X"].apply(limpar_e_padronizar_telefone)
```

```python
df.head(30)
```

```python
df.to_csv(saida_csv, index=False)
print(f"Arquivo CSV salvo com sucesso: {saida_csv}")

df.to_excel(saida_excel, index=False)
print(f"Arquivo Excel salvo com sucesso: {saida_excel}")
```
